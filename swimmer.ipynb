# Swimmer Experiments
This notebook contains the results of various algorithms on Swimmer dataset.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.genfromtxt('swimmer.txt')
S_mat = np.genfromtxt('S.txt')
W = np.genfromtxt('W.txt')
H = np.genfromtxt('H.txt')
L = np.genfromtxt('L.txt')
S = S_mat.reshape(1024, 256, 17)
#S = S_mat.reshape(8, 10, 3)
S_sum = np.sum(S, axis=2)

n_components = 17
matplotlib.rcParams['figure.figsize'] = [15, 10]
```

```python
Y = X
Y = W@H
plt.imshow(Y[:, 8].reshape(32, 32), cmap='gray')
```

```python
fig = plt.figure()
plt.imshow(S_sum[:, 99].reshape(32, 32), cmap='gray')
plt.xticks([])
plt.yticks([])
#fig.savefig('figure/swimmer_99.pdf', format='pdf', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(24, 16))
y = 99
for i in range(n_components):
    plt.subplot(3, 6, i + 1)
    plt.imshow(S[:, y, i].reshape(32, 32), cmap='gray')
    plt.title(r'$S_{:, ' + str(y) + ', ' + str(i) + r'}$')
    plt.xticks([])
    plt.yticks([])
    
#fig.savefig('figure/swimmer_bld_mult_S.pdf', format='pdf', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(24, 8))
plt.imshow(H[:, 90:120 ], cmap='gray')
#fig.savefig('figure/swimmer_nmf_H.pdf', format='pdf', bbox_inches='tight')
```

```python
fig = plt.figure(figsize=(24, 16))
for i in range(n_components):
    plt.subplot(3, 6, i + 1)
    plt.imshow(W[:, i].reshape(32, 32), cmap='gray')
    plt.title(r'$W_{:' + str(i) + r'}$')
    plt.xticks([])
    plt.yticks([])
```
