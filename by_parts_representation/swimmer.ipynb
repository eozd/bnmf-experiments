```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

## Read Matrices/Tensors

```python
# Read matrices and tensors factorized by the inference algorithm

X = np.genfromtxt('swimmer.txt')
S_mat = np.genfromtxt('S.txt')
W = np.genfromtxt('W.txt')
H = np.genfromtxt('H.txt')
L = np.genfromtxt('L.txt')
S = S_mat.reshape(1024, 256, 17)
S_sum = np.sum(S, axis=2)

n_components = 17
matplotlib.rcParams['figure.figsize'] = [15, 10]
```

## Plot Samples From Swimmer

```python
fig = plt.figure()
plt.imshow(S_sum[:, 99].reshape(32, 32), cmap='gray')
plt.xticks([])
plt.yticks([]);
fig.savefig('figure/swimmer_99.eps', format='eps', bbox_inches='tight')
```

```python
fig = plt.figure()
plt.imshow(S_sum[:, 45].reshape(32, 32), cmap='gray')
plt.xticks([])
plt.yticks([]);
fig.savefig('figure/swimmer_45.eps', format='eps', bbox_inches='tight')
```

## Plot Discovered Components
There are 17 components:
* 4 components (every rotated version is a component) per each arm/leg for a total of 16
* Body for a total of 1

```python
fig = plt.figure(figsize=(24, 16))
for i in range(n_components):
    plt.subplot(3, 6, i + 1)
    plt.imshow(W[:, i].reshape(32, 32), cmap='gray')
    plt.title(r'$W_{:' + str(i) + r'}$', fontsize=32)
    plt.xticks([])
    plt.yticks([])
    
fig.savefig('figure/swimmer_W.eps', format='eps', bbox_inches='tight')
```
